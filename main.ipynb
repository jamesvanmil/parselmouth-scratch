```python
import parselmouth

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
```

```python
sns.set() # Use seaborn's default style to make attractive graphs
plt.rcParams['figure.dpi'] = 200 # Show nicely large images in this notebook
```

```python
sounds = glob("sound/*.mp3")
print(sounds)
```

```python
for sound in sounds:
    print(sound)
    
    snd = parselmouth.Sound(sound)
    pitch = snd.to_pitch(time_step=(1/60))
    pitch_values = pitch.selected_array['frequency']
    pitch_values[pitch_values==0] = np.nan
    plt.xlabel("time [s]")
    plt.ylabel("frequency")
    plt.plot(pitch.xs(), pitch_values, 'o', markersize=1)
    
    plt.show()
    
    intensity = snd.to_intensity(time_step=(1/60))
    intensity_values = intensity.values.T
    intensity_values[intensity_values==0] = np.nan
    plt.xlabel("time [s]")
    plt.ylabel("intensity")
    plt.plot(intensity.xs(), intensity_values, 'o', markersize=1)
    
    plt.show()
    
    print(len(pitch_values))
```

```python
plt.figure()
plt.plot(snd.xs(), snd.values.T)
plt.xlim([snd.xmin, snd.xmax])
plt.xlabel("time [s]")
plt.ylabel("amplitude")
plt.show() # or plt.savefig("sound.png"), or plt.savefig("sound.pdf")
```
